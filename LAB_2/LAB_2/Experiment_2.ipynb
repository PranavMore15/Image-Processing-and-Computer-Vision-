```python
#Step 1: Import Required Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Step 2: Load Image

img_color = cv2.imread("/content/drive/MyDrive/IPCV/LAB_2/House.jpg")


# Convert BGR to RGB for correct display
img_rgb = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

# Display Image
plt.imshow(img_rgb)
plt.title("Original Image (Color)")
plt.axis('off')
plt.show()

```

```python
 # Step 3: Convert to Grayscale
 # Grayscale conversion is necessary because image enhancement formulas such as negative, contrast stretching, bit plane slicing, and gray level slicing are defined for single-channel intensity images

 gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
```

```python
# Step 4: Image Negative
negative = 255 - gray

plt.imshow(negative, cmap='gray')
plt.title("Image Negative")
plt.axis('off')
plt.show()
```

```python
# Step 5: Contrast Stretching
r_min = np.min(gray)
r_max = np.max(gray)

contrast_stretched = ((gray - r_min) / (r_max - r_min)) * 255
contrast_stretched = contrast_stretched.astype(np.uint8)

plt.imshow(contrast_stretched)
plt.title("Contrast Stretched Image")
plt.axis('off')
plt.show()
```

```python
# Step 6: Bit Plane Slicing
plt.figure(figsize=(10,6))

for i in range(8):
    bit_plane = (gray >> i) & 1
    plt.subplot(2,4,i+1)
    plt.imshow(bit_plane, cmap='gray')
    plt.title(f'Bit Plane {i}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Step 7: Gray Level Slicing
# With Background Suppressed
lower = 100
upper = 150

gray_slice = np.zeros_like(gray)
gray_slice[(gray >= lower) & (gray <= upper)] = 255

plt.imshow(gray_slice, cmap='gray')
plt.title("Gray Level Slicing (Background Suppressed)")
plt.axis('off')
plt.show()
```

```python
# Without Background Suppressed
gray_slice2 = gray.copy()
gray_slice2[(gray >= lower) & (gray <= upper)] = 255

plt.imshow(gray_slice2, cmap='gray')
plt.title("Gray Level Slicing (Background Retained)")
plt.axis('off')
plt.show()
```
